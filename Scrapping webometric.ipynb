```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.webometrics.info/en/Asia_es/Indonesia'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36 Edg/123.0.0.0',
}
req = requests.get(url, headers=headers)
soup = BeautifulSoup(req.text, 'html.parser')
soup
```

```python
first_row = soup.find_all('th')
first_row 
```

```python
titles = [title.text for title in first_row] #strip() untuk menghapus spasi
titles
```

```python
df = pd.DataFrame(columns = titles)
df
```

```python
review = soup.findAll('tr')
review
```

```python
individual_row_data = [data.text.strip() for data in review]
individual_row_data

```

```python
for row in review[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```

```python
df.to_csv('universities.csv', index=False)
```
