```python
url = "https://www.topuniversities.com/world-university-rankings/2023?region=Asia&countries=id"

response = requests.get(url)
soup = bs4(response.content, "html.parser")

# Cari semua elemen yang memiliki class "rankings-indicator" (berisi informasi mengenai skor)
score_elements = soup.find_all("div", class_="rankings-indicator")

# Loop melalui setiap elemen dan ambil teksnya
for score_element in score_elements:
    # Jika teksnya mengandung "Overall Score", kita ambil nilai skornya
    if "Overall Score" in score_element.get_text():
        overall_score = score_element.find_next("div", class_="indicator-score").get_text().strip()
        print(f"Overall Score: {overall_score}")
        break  # Keluar dari loop setelah mendapatkan skor overall
```
