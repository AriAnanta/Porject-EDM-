```python
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup as bs4
```

```python
page = requests.get("https://databoks.katadata.co.id/datapublish/2023/06/21/10-ptn-dengan-pendaftar-terbanyak-di-snbt-2023-universitas-brawijaya-teratas").text
page
```

```python
page_soup = bs4(page, "html.parser")
divs = page_soup.find_all('ol', {"class":"ol-number"})
divs
```

```python
target = [divs.text for divs in divs]
target
```

```python
cleaned_data = [data.strip() for data in target[0].split('\n') if data.strip() != '']
universities = [data.split(':')[0] for data in cleaned_data]
applicants = [int(data.split(':')[1].split(' ')[1].replace('pendaftar', '').replace('.', '')) for data in cleaned_data]

df = pd.DataFrame({
    'Nama Universitas': universities,
    'Jumlah Pendaftar': applicants
})
df
```

```python
df.to_csv('data_pendaftarPTN_databoks.csv', index=False)
```
